```python
# Task 1: Assigning Values to Variables

# Create an integer variable
my_integer = 5

# Create a floating-point variable
my_float = 3.14

# Create a string variable
my_string = "Python"

# Print all three variables on one line, separated by commas
print(my_integer, my_float, my_string)


# Task 2: Data Types

# Declare a variable 'age' and assign an integer value
age = 18

# Print the data type of 'age'
print(type(age))

# Change 'age' to a floating-point number
age = 18.0

# Print the data type again
print(type(age))

# Declare a variable 'is_student' and assign a boolean value
is_student = True

# Print the data type of 'is_student'
print(type(is_student))


# Task 3: Basic Input and Output

# Ask the user to enter their name
name = input("Enter your name: ")

# Print a message saying Hello to the user
print("Hello,", name + "!")

# Ask the user for their age
user_age = int(input("Enter your age: "))

# Check if the user is eligible to vote (18 years or older)
if user_age >= 18:
    print("You are eligible to vote.")
else:
    print("You are not eligible to vote.")


# Task 4: Converting Data Types

# Ask the user to enter a number as a string
number_str = input("Enter a number: ")

# Convert the string to an integer
number_int = int(number_str)

# Print the square of the number
print(number_int * number_int)

# Create a string that represents a floating-point number
float_str = "4.56"

# Convert the string to a float
float_num = float(float_str)

# Multiply the float by 2 and print the result
print(float_num * 2)


# Task 5: Arithmetic Operators

# Add 45 and 67
print(45 + 67)

# Subtract 89 from 156
print(156 - 89)

# Multiply 12 by 7
print(12 * 7)

# Divide 100 by 4
print(100 / 4)

# Find the remainder when 10 is divided by 3
print(10 % 3)

# Raise 3 to the power of 4
print(3 ** 4)


# Task 6: Comparison Operators

# Create two numbers
a = 10
b = 20

# Compare the two numbers
print(a == b)  # Equal
print(a != b)  # Not equal
print(a < b)   # Less than
print(a > b)   # Greater than
print(a <= b)  # Less than or equal to
print(a >= b)  # Greater than or equal to

# Ask the user to enter two numbers
num1 = int(input("Enter the first number: "))
num2 = int(input("Enter the second number: "))

# Check if the first number is greater than or equal to the second
if num1 >= num2:
    print("First number is greater than or equal to second number.")
else:
    print("First number is less than second number.")


# Task 7: Logical Operators

# Ask the user to enter a number
number = int(input("Enter a number: "))

# Check if the number is greater than 10 and less than 20
if number > 10 and number < 20:
    print("The number is between 10 and 20.")
else:
    print("The number is not between 10 and 20.")

# Ask the user to enter another number
number = int(input("Enter a number: "))

# Check if the number is less than 5 or greater than 50
if number < 5 or number > 50:
    print("The number is less than 5 or greater than 50.")
else:
    print("The number is between 5 and 50.")


# Task 8: Conditional Statements

# Ask the user for their age
age = int(input("Enter your age: "))

# Check if the user is an adult
if age >= 18:
    print("You are an adult.")
else:
    print("You are a minor.")

# Ask the user for a number
number = int(input("Enter a number: "))

# Check divisibility
if number % 3 == 0 and number % 5 == 0:
    print("FizzBuzz")
elif number % 3 == 0:
    print("Fizz")
elif number % 5 == 0:
    print("Buzz")
else:
    print(number)


# Task 9: Loops

# For loop to print numbers from 1 to 10
for i in range(1, 11):
    print(i)

# While loop to print even numbers from 2 to 20
i = 2
while i <= 20:
    print(i)
    i = i + 2

# Ask the user for a number n
n = int(input("Enter a number: "))

# Print numbers from 1 to n that are divisible by 4
for i in range(1, n + 1):
    if i % 4 == 0:
        print(i)
```
