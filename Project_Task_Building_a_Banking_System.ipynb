```python
# Set up initial username, password, and number of allowed tries
username = "user1"
password = "pass123"
tries = 3

# Loop for user login attempts
# This 'while' loop will continue running as long as 'tries' is greater than 0
while tries > 0:
    user_input = input("Enter username: ")  # Ask user to enter username
    pass_input = input("Enter password: ")  # Ask user to enter password

    if user_input == username and pass_input == password:
        # IF the entered username and password match the correct ones:
        print("Login successful!")
        break  # Exit the login loop immediately if login is successful
    else:
        # ELSE, meaning username/password was wrong
        tries = tries - 1  # Decrease the number of tries left
        print("Wrong username or password!")
        print("Tries left:", tries)

# 'if tries == 0' is OUTSIDE the while loop because:
# After exiting the login loop (either because of successful login or tries running out),
# we need to check whether the user failed all login attempts.
if tries == 0:
    print("Too many wrong tries Program will stop!")  # Inform the user and end program here

# 'if tries != 0' means login was successful
# If the user logged in successfully, they can now access the main banking menu
if tries != 0:
    balance = 1000  # Initialize account balance

    # 'while True' creates an INFINITE loop.
    # This means the menu will keep appearing again and again
    # until we manually break the loop (when the user chooses 'Exit').
    while True:  # Main menu loop
        print("")
        print("----- MAIN MENU -----")
        print("1. Check balance")
        print("2. Deposit money")
        print("3. Withdraw money")
        print("4. Exit")

        choice = input("Please choose 1, 2, 3, or 4: ")  # Get user's choice

        if choice == "1":
            # Option 1: Display current balance
            print("Your balance is:", balance)

        elif choice == "2":
            # Option 2: Deposit money into account
            deposit = input("How much do you want to deposit?: ")
            deposit = float(deposit)  # Convert input to float
            if deposit > 0:
                balance = balance + deposit  # Update balance
                print("Money added! Your new balance is:", balance)
            else:
                print("Please enter a number more than 0!")

        elif choice == "3":
            # Option 3: Withdraw money from account
            withdraw = input("How much do you want to withdraw?: ")
            withdraw = float(withdraw)  # Convert input to float
            if withdraw > 0 and withdraw <= balance:
                balance = balance - withdraw  # Update balance
                print("Money taken out! Your new balance is:", balance)
            else:
                print("Not enough money or bad number!")

        elif choice == "4":
            # Option 4: Exit the program
            print("Goodbye! Thank you for using the bank!")
            break  # 'break' exits the infinite 'while True' loop and ends the program

        else:
            # If an invalid menu option is selected
            print("That is not a choice! Please pick 1, 2, 3, or 4.")

# Additional Explanation:
# - The 'while True' makes the menu keep looping infinitely after every operation.
# - Only when the user chooses option 4 ("Exit"), the 'break' command runs,
#   breaking out of the infinite loop and stopping the program.
```
